# Porcentaje por departamento de tipo de cultivo

```python
#resultados del gobierno
gobierno_df = pd.read_csv("../estimaciones/estimaciones_inta.csv",encoding="latin-1")
gob_df = gobierno_df.groupby("nombre").agg(
    superficie_total =  ('ha', 'sum'))
gob_df.reset_index(inplace=True)
gob_final = pd.merge(gob_df, gobierno_df, on="nombre", how="left")
gob_final["porcentaje_cubierto_gob"] = (gob_final["ha"] / gob_final["superficie_total"])*100
```

```python
umbrales = [0.3,0.4,0.56,0.6, 0.65,0.7,0.8]
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
#umbrales = [0.6]

mapes = []
modelos = []
mse_list = []

for umbral in umbrales:
    results= f"../model/threshhold_{umbral}/randomforest_iterations_deptos_prediction.csv"
    results_df = pd.read_csv(results)
    print(results_df)
    res_df = results_df.groupby("nombre").agg(
        superficie_total =  ('ha', 'sum'))

    res_df.reset_index(inplace=True)
    res_final = pd.merge(res_df, results_df, on="nombre", how="left")
    res_final["porcentaje_cubierto"] = (res_final["ha"] / res_final["superficie_total"])*100

    final = pd.merge(res_final[["nombre","cultivo","porcentaje_cubierto",]], gob_final[["porcentaje_cubierto_gob","cultivo","nombre"]], left_on=["nombre","cultivo"], right_on = ["nombre","cultivo"], how="left")

    final.dropna(inplace=True)


    modelos.append(umbral)
    mape  = mean_absolute_percentage_error(final["porcentaje_cubierto"], final["porcentaje_cubierto_gob"])
    mapes.append(mape)

    mse= mean_squared_error(final["porcentaje_cubierto"], final["porcentaje_cubierto_gob"])
    mse_list.append(mse)

evaluaciones = pd.DataFrame({"modelo":modelos,"mape":mapes,"mse":mse_list})
evaluaciones
```

```python
thresholds = [0.3, 0.4, 0.56]
cant_df = pd.DataFrame()
per_df = pd.DataFrame()

df_cantidades = []
df_per = []
for umbral in thresholds:
    folder_path = f"../model/threshhold_{umbral}/"

    iteraciones = glob(f'../model/threshhold_{umbral}/randomforest_iterations_*/')
    print(sorted(iteraciones))
    for i, iter in enumerate(iteraciones):
        probabilidades = glob(iter+"/*.npy")
        for prob in probabilidades:
            data = np.load(f"{prob}")
            max_df = pd.DataFrame(data.max(axis=1), columns=['max'])
            umbral2 = str(umbral)
            clase_df = pd.DataFrame(data.argmax(axis=1), columns=[f"clase_iter_{umbral2}_{i}"])
            preds = pd.merge(max_df, clase_df, left_index=True, right_index=True)
            cant_total = len(preds)
            nuevos = preds[preds["max"] > 0.4]
            cant_nuevos = len(nuevos)
            cantidad = nuevos[f"clase_iter_{umbral2}_{i}"].value_counts()
            porcentaje = nuevos[f"clase_iter_{umbral2}_{i}"].value_counts(normalize=True)
            print(f"La cantidad de predicciones nuevas son {cant_nuevos} que representa un total de {cant_nuevos/cant_total}")
            cant_df[f"iter_{umbral2}__{i}"] = cantidad
            per_df[f"iter_{umbral2}__{i}"] = porcentaje
            df_cantidades.append(cant_df)
            df_per.append(per_df)



```
